```python
from math import log, exp
import numpy as np
from scipy.optimize import line_search
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def binary_search(f, a, b, epsilon, callback=None):
    c = (a + b) / 2.0
    while abs(b - a) > epsilon:
#         Check left subsegment
        y = (a + c) / 2.0
        if f(y) <= f(c):
            b = c
            c = y
        else:
#         Check right subsegment
            z = (b + c) / 2.0
            if f(c) <= f(z):
                a = y
                b = z
            else:
                a = c
                c = z
        if callback is not None:
            callback(a, b)
    return c
```

```python

def my_callback(a, b, left_bound, right_bound, approximation):
    left_bound.append(a)
    right_bound.append(b)
    approximation.append((a + b) / 2.0)
```

```python
import numpy as np

left_boud_bs = []
right_bound_bs = []
approximation_bs = []

callback_bs = lambda a, b: my_callback(a, b, 
            left_boud_bs, right_bound_bs, approximation_bs)

# Target unimodal function on given segment
f = lambda x: (x - 2) * x * (x + 2)**2 # np.power(x+2, 2)
# f = lambda x: -np.sin(x)
x_true = -2
# x_true = np.pi / 2.0
a = -3
b = -1.5
epsilon = 1e-8
x_opt = binary_search(f, a, b, epsilon, callback_bs)
print(np.abs(x_opt - x_true))
plt.figure(figsize=(10,6))
plt.plot(np.linspace(a,b), f(np.linspace(a,b)))
plt.title("Objective function", fontsize=28)
plt.xticks(fontsize = 28)
_ = plt.yticks(fontsize = 28)
```

```python
def golden_search(f, a, b, tol=1e-5, callback=None):
    tau = (np.sqrt(5) + 1) / 2.0
    y = a + (b - a) / tau**2
    z = a + (b - a) / tau
    while b - a > tol:
        if f(y) <= f(z):
            b = z
            z = y
            y = a + (b - a) / tau**2
        else:
            a = y
            y = z
            z = a + (b - a) / tau
        if callback is not None:
            callback(a, b)
    return (a + b) / 2.0
```

```python
left_boud_gs = []
right_bound_gs = []
approximation_gs = []

cb_gs = lambda a, b: my_callback(a, b, left_boud_gs, right_bound_gs, approximation_gs)
x_gs = golden_search(f, a, b, epsilon, cb_gs)

print(f(x_opt))
print(f(x_gs))
print(np.abs(x_opt - x_true))
```

$$
H_{k+1} = H_k - \frac{H_ky_k y_k^{\top}H_k}{y^{\top}_kH_ky_k} + \frac{s_ks^{\top}_k}{y^{\top}_ks_k}

$$
u^{\top}H_{k+1}u = u^{\top}H_ku - \frac{u^{\top}H_ky_k y_k^{\top}H_ku}{y^{\top}_kH_ky_k} + \frac{u^{\top}s_ks^{\top}_ku}{y^{\top}_ks_k}

```python
from math import log, exp
import numpy as np
from scipy.optimize import line_search
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
f = lambda x: log(exp(x) + exp(-x))
gradf = lambda x: 1 - 2/ (exp(2*x) + 1)
hessf = lambda x: 4 / (exp(x) + exp(-x))**2
```

```python
def Newton(f, gradf, hessf, x0, line_search=None, epsilon=1e-6, num_iter=1000, 
                    disp=False, callback=None, **kwargs):
    x = x0
    iteration = 0
    opt_arg = {"f": f, "grad_f": gradf}
    for key in kwargs:
        opt_arg[key] = kwargs[key]
    res = {"x":[],"num_iter":[],"tol":[]}
    while True:
        gradient = gradf(x)
        hess = hessf(x)
        if len(x) >= 2:
            h = np.linalg.solve(hess, -gradient)
        else:
            h = -gradient/hess
            
        #search alpha:
        if line_search is None:
            alpha = 1
        else:
            search_alpha = line_search(f, gradf, x, h)
            alpha = search_alpha[0]
            if alpha is None:
                alpha = 1
                
        x = x + alpha * h
        if callback is not None:
            callback(x)
        iteration += 1
        
        if len(x) >= 2:
            tol = np.linalg.norm(gradient, 2)
        else:
            tol = abs(gradient)
        
        res["x"].append(x)
        res["num_iter"].append(iteration)
        res["tol"].append(tol)
        
        if disp:
            print("Current function val =", f(x))
            print("Current gradient norm = ", np.linalg.norm(gradf(x)))

        if tol < epsilon:
            break
        if iteration >= num_iter:
            break
    return res
```

```python
res1 = Newton(f, gradf, hessf, np.array([1]), line_search = line_search, num_iter=10)
res2 = Newton(f, gradf, hessf, np.array([1]), num_iter=10)
res3 = Newton(f, gradf, hessf, np.array([1.1]), line_search = line_search, num_iter=10)
res4 = Newton(f, gradf, hessf, np.array([1.1]), num_iter=5)
print(res1["tol"], "\n", res2["tol"])
```

```python
plt.figure(figsize=(12,10))
fontsize = 26

plt.xlabel(r"Number of iteration, $k$", fontsize=20)
plt.ylabel(r"Convergence rate", fontsize=20)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)

plt.semilogy(res1["tol"], label=r"line search, $x_0 = 1$")
plt.semilogy(res2["tol"], label=r"$\alpha$ = 1, $x_0 = 1$")
plt.semilogy(res3["tol"], label=r"line search, $x_0 = 1.1$")
plt.semilogy(res4["tol"], label=r"$\alpha$ = 1, $x_0 = 1.1$")

plt.legend(fontsize=fontsize, loc="best")
```

Из точки 1 мметод сходится и сходимость не зависит от выбор шага- графики совпали. Из точки 1.1 при постоянном шаге метод расходится, но сходится при демпфированном методе Ньютона. Разницу можно объяснить тем, что можно выбрать шаг меньше 1 и не перескочить нужную точку.

```python
%timeit res = Newton(f, gradf, hessf, np.array([1]), line_search = line_search, num_iter=10)
%timeit res = Newton(f, gradf, hessf, np.array([1]), num_iter=10)
%timeit res = Newton(f, gradf, hessf, np.array([1.1]), line_search = line_search, num_iter=10)
%timeit res = Newton(f, gradf, hessf, np.array([1.1]), num_iter=5)
```

Демпфированный метод Ньютона работает дольше, но увеличивает область сходимости

```python
def bfgs_method(f, gradf, x0, maxiter=1000, epsilon=10e-3):

    k = 0
    grad = gradf(x0)
    N = len(x0)
    I = np.eye(N, dtype=int)
    Hk = I
    x = x0
    res = {"x":[],"num_iter":[],"tol":[]}
   
    while np.linalg.norm(grad) > epsilon and k < maxiter:
        h = -np.dot(Hk, grad)
        
        search_alpha = line_search(f, gradf, x, h)
        alpha_k = search_alpha[0]
        if alpha_k is None:
            alpha_k = 1
            
        x_new = x + alpha_k * h
        sk = x_new - x
        x = x_new
        
        grad_new = gradf(x_new)
        yk = grad_new - grad
        grad = grad_new
        
        k += 1
        res["x"].append(x)
        res["num_iter"].append(k)
        res["tol"].append(np.linalg.norm(grad, 2))
        
        ro = 1.0 / (np.dot(yk, sk))
        Hk_yk = Hk.dot(yk)
        A1 = I - ro * sk[:, np.newaxis] * yk[np.newaxis, :]
        A2 = I - ro * yk[:, np.newaxis] * sk[np.newaxis, :]
        Hk = np.dot(A1, np.dot(Hk, A2)) + (ro * sk[:, np.newaxis] * sk[np.newaxis, :])
    return res
```
