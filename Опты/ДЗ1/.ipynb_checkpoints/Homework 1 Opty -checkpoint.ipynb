```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import numpy as np
from tqdm.auto import tqdm
import scipy.optimize
import math
```

```python
def binary_search(f, a, b, epsilon, callback=None):
    c = (a + b) / 2.0
    while abs(b - a) > epsilon:
#         Check left subsegment
        y = (a + c) / 2.0
        if f(y) <= f(c):
            b = c
            c = y
        else:
#         Check right subsegment
            z = (b + c) / 2.0
            if f(c) <= f(z):
                a = y
                b = z
            else:
                a = c
                c = z
        if callback is not None:
            callback(a, b)
    return c
```

```python
def my_callback(a, b, left_bound, right_bound, approximation):
    left_bound.append(a)
    right_bound.append(b)
    approximation.append((a + b) / 2.0)
```

```python
def golden_search(f, a, b, tol=1e-5, callback=None):
    tau = (np.sqrt(5) + 1) / 2.0
    y = a + (b - a) / tau**2
    z = a + (b - a) / tau
    while b - a > tol:
        if f(y) <= f(z):
            b = z
            z = y
            y = a + (b - a) / tau**2
        else:
            a = y
            y = z
            z = a + (b - a) / tau
        if callback is not None:
            callback(a, b)
    return (a + b) / 2.0
```

```python
left_boud_gs = []
right_bound_gs = []
approximation_gs = []

cb_gs = lambda a, b: my_callback(a, b, left_boud_gs, right_bound_gs, approximation_gs)
x_gs = golden_search(f, a, b, epsilon, cb_gs)

print(f(x_opt))
print(f(x_gs))
print(np.abs(x_opt - x_true))
```

### Метод золотого сечения

```python
def get_y(a, b):
    return a + (2/(3 + np.sqrt(5)))*(b-a)

def get_z(a, b):
    return a + ((np.sqrt(5) - 1)/2)*(b-a)

def gold(f, a, b, k_max, write = None):
    y = get_y(a, b)
    z = get_z(a, b)
    for k in range(k_max):
        if f(y) <= f(z):
            b = z
            z = y
            y = get_y(a, b)
        else:
            a = y
            y = z
            z = get_z(a, b)
        if write is not None:
            write.append(f((a+b)/2))
    return f((a+b)/2)
```

```python
gold(f, a, b, k_max)
```

### Метод Фибоначчи

```python
fibonacci = [0, 1]
for i in range(2, 20):
    fibonacci.append(fibonacci[i-2] + fibonacci[i-1])
```

```python
def get_l(k_max, k, a, b):
    return a + (fibonacci[k_max - k - 1]/fibonacci[k_max - k + 1])*(b-a)

def get_m(k_max, k, a, b):
    return a + (fibonacci[k_max - k]/fibonacci[k_max - k + 1])*(b-a)

def fib(f, a, b, k_max, write = None):
    k_max += 1
    l = get_l(k_max, 1, a, b)
    m = get_m(k_max, 1, a, b)
    for k in range(2, k_max + 1):
        if f(l) <= f(m):
            b = m
            m = l
            l = get_l(k_max, k, a, b)
        else:
            a = l
            l = m
            m = get_m(k_max, k, a, b)
        if write is not None:
            write.append(f((a+b)/2))
        
    x = (a+b)/2
    return f(x)
```

```python
def grad_f(x):
    return (pow(math.e,2*x)-1)/(pow(math.e,2*x)+1)
```

```python
def hess_f(x):
    return (4*pow(math.e,2*x))/(pow(math.e,2*x)+1)**2
```

```python
def next_x(x):
    return x - grad_f(x)/hess_f(x)
```


### Демпфированный метод Ньютона (с переменным шагом)
можно было найти альфу аналитически (одномерный поиск), $$
\alpha_{k}=\arg \min _{\alpha>0} f\left(x_{k}-\alpha\left[\nabla^{2} f\left(x_{k}\right)\right]^{-1} \nabla f\left(x_{k}\right)\right)
$$ тогда метод сошелся бы за 1 шаг, что показано 
письменно на листочке

а можно исползовать дробление, начиная $$
\alpha=1, \alpha_{k}=\gamma \alpha_{k}, 0<\gamma<1
$$
до выполнения каких-то условий, например

$$
\begin{array}{c}f\left(x_{k+1}\right) \leq f\left(x_{k}\right)-\alpha q\left(\left[\nabla^{2} f\left(x_{k}\right)\right]^{-1} \nabla f\left(x_{k}\right), \nabla f\left(x_{k}\right)\right) \\ \left\|\nabla f\left(x_{k+1}\right)\right\|^{2} \leq(1-\alpha q)\left\|\nabla f\left(x_{k}\right)\right\|^{2} \\ 0<q<1\end{array}
$$

```python
def d_next_x(x, a):
    a *= 1/2
    x = x - a*grad_f(x)/hess_f(x)
    return x, a
```

```python
x = 1.1
a = 1
x_d = [x]
for i in range(k):
    x, a = d_next_x(x, a)
    x_d.append(x)
    print(x)
```
